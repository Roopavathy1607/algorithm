```python
import random
import time
import matplotlib.pyplot as plt
import heapq
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
def heap_sort(arr):
    heapq.heapify(arr)
    sorted_arr = []
    while arr:
        sorted_arr.append(heapq.heappop(arr))
    return sorted_arr
def measure_time(sort_func, arr):
    start_time = time.time()
    sort_func(arr)
    return time.time() - start_time
def experiment(max_n, step=100):
    ns = []
    insertion_times = []
    heap_times = []

    for n in range(step, max_n+1, step):
        arr = random.sample(range(1, 10*n), n)
        insertion_time = measure_time(insertion_sort, arr.copy())
        insertion_times.append(insertion_time)
        heap_time = measure_time(heap_sort, arr.copy())
        heap_times.append(heap_time)

        ns.append(n)

    return ns, insertion_times, heap_times
max_n = 2000
ns, insertion_times, heap_times = experiment(max_n)
plt.figure(figsize=(10, 6))
plt.plot(ns, insertion_times, label='Insertion Sort', color='blue', marker='o')
plt.plot(ns, heap_times, label='Heap Sort', color='red', marker='x')

plt.title('Time Complexity of Insertion Sort and Heap Sort')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
